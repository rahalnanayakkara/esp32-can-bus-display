```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("main/Lambda.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
THRESH = 100

img[img>THRESH] = 255
img[img<=THRESH] = 0

plt.imshow(img)
```

```python
x = img.reshape(-1).astype(np.uint16)
```
